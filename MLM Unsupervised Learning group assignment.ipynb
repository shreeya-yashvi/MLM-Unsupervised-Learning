### Runtime
- **Agglomerative Clustering**:
  - Mean runtime: 0.8448 seconds  
  - Standard deviation: 0.0509 seconds  
  - **Observation**:  
    - Slightly higher mean runtime and variability due to the hierarchical computation of the linkage matrix and pairwise comparisons.

- **K-Means**:
  - Mean runtime: 0.7733 seconds  
  - Standard deviation: 0.0280 seconds  
  - **Observation**:  
    - Faster and more consistent performance due to its iterative centroid updating process.

### Memory Usage
- **Agglomerative Clustering**:
  - Mean memory usage: 107.44 MB  
  - Peak memory usage: 107.57 MB  
  - **Observation**:  
    - Agglomerative Clustering is highly memory-intensive due to the need to store the distance matrix between all data points, which grows quadratically with the number of data points (O(n²)).

- **K-Means**:
  - Mean memory usage: 0.42 MB  
  - Peak memory usage: 0.56 MB  
  - **Observation**:  
    - K-Means exhibits extremely low memory usage because it primarily stores the cluster centers. This is particularly efficient when the number of clusters (k) is small compared to the number of data points (n).

### Overall Analysis
1. **Memory Efficiency**:  
   - K-Means is vastly more memory-efficient than Agglomerative Clustering, making it suitable for large datasets that might exceed memory capacity with Agglomerative Clustering.

2. **Runtime Performance**:  
   - K-Means is slightly faster than Agglomerative Clustering in this case. The difference in runtime tends to grow with larger datasets.

3. **Scalability**:  
   - K-Means scales much better to large datasets due to its lower memory requirements and faster runtime.  
   - Agglomerative Clustering is better suited for smaller datasets or scenarios where the hierarchical structure of clusters is critical.

```python
from scipy.stats import chi2_contingency

# Prepare a contingency table for a nominal and ordinal variable
# Using "Shipping_Method" (nominal) and "Date" (ordinal) as an example
contingency_table = pd.crosstab(sy45_ss["Shipping_Method"], sy45_ss["Date"])

# Perform the Chi-Square test of homogeneity
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Present the results
{
    "Chi-Square Statistic": chi2_stat,
    "p-value": p_value,
    "Degrees of Freedom": dof,
    "Expected Frequencies": expected
}
```

- **Degrees of Freedom (DoF)**: 3430  
- **Expected Frequencies**: Matrix of expected counts for each combination of categories (not fully displayed here).

## Interpretation:
- The **p-value** of 0.4124 indicates no significant difference in the distribution of "Shipping_Method" across the "Date" categories.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

def visualize_kmeans_pca(data_subset, cluster_labels, title="K-Means Clustering Results (PCA)"):
    """
    Visualizes K-Means clustering results using PCA for dimensionality reduction.

    Args:
        data_subset: The original data subset as a Pandas DataFrame.
        cluster_labels: The cluster labels from the K-Means model.
        title: The title of the plot.

    Returns:
        None (displays the plot).
    """
    try:
        # 1. Data Scaling (Crucial for PCA)
        scaler = StandardScaler()
        scaled_data = scaler.fit_transform(data_subset)

        # 2. Apply PCA to reduce to 2 dimensions
        pca = PCA(n_components=2)
        pca_result = pca.fit_transform(scaled_data)

        # 3. Create a DataFrame with PCA results and cluster assignments
        pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2'])
        pca_df['cluster'] = cluster_labels

        # 4. Explained Variance Ratio
        explained_variance = pca.explained_variance_ratio_
        print(f"Explained Variance Ratio: {explained_variance}")
        print(f"Total Explained Variance: {sum(explained_variance)}")

        # 5. Plot the PCA results
        plt.figure(figsize=(8, 6))
        sns.scatterplot(x='PC1', y='PC2', hue='cluster', data=pca_df, palette=['red', 'blue', 'green', 'purple', 'orange', 'yellow', 'brown', 'pink','gray', 'olive'])  # Customizable palette
        plt.title(title)
        plt.xlabel(f'Principal Component 1 ({explained_variance[0]*100:.1f}%)')
        plt.ylabel(f'Principal Component 2 ({explained_variance[1]*100:.1f}%)')
        plt.show()

    except ValueError as e:
        print(f"Error visualizing K-Means PCA: {e}")
        print("Check if the lengths of data_subset and cluster_labels match, and if data_subset has numeric columns.")
    except Exception as e:
        print(f"An unexpected error occurred: {e}")


#Example usage (assuming you have your data in sy45_finalsample_subset and cluster labels in km_2cluster_model):
visualize_kmeans_pca(sy45_finalsample_subset[['Value', 'Weight', 'Category']], sy45_km_2cluster_model, title="K-Means with 2 Clusters (PCA)")

```

```python
import time
import numpy as np
from sklearn.cluster import KMeans

def time_model_run(model, data):
    start_time = time.time()
    model.fit(data)
    end_time = time.time()
    return end_time - start_time

# Example for KMeans
kmeans_model = KMeans(n_clusters=2, init='random', random_state=4546)

# Run multiple times for statistics
num_runs = 10
runtimes = [time_model_run(kmeans_model, sy45_finalsample_subset) for _ in range(num_runs)]

mean_runtime = np.mean(runtimes)
std_runtime = np.std(runtimes)

print(f"KMeans runtime statistics over {num_runs} runs:")
print(f"Mean runtime: {mean_runtime:.4f} seconds")
print(f"Standard deviation: {std_runtime:.4f} seconds")
```

## Memory Used

### KMeans Model

```python
import tracemalloc
import numpy as np
from sklearn.cluster import KMeans

def get_model_memory_usage(model, data):
    tracemalloc.start()
    model.fit(data)
    current, peak = tracemalloc.get_traced_memory()
    tracemalloc.stop()
    return peak / (1024 * 1024)  # Convert bytes to MB

# Example for KMeans
kmeans_model = KMeans(n_clusters=2, init='random', random_state=4546)

# Run multiple times for statistics
num_runs = 10
memory_usages = [get_model_memory_usage(kmeans_model, sy45_finalsample_subset) for _ in range(num_runs)]

mean_memory_usage = np.mean(memory_usages)
max_memory_usage = np.max(memory_usages)

print(f"KMeans memory usage statistics over {num_runs} runs:")
print(f"Mean memory usage: {mean_memory_usage:.2f} MB")
print(f"Peak memory usage: {max_memory_usage:.2f} MB")
```
